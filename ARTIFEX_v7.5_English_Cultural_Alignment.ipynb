# 🤖 MODEL_RESPONSES: SIMULATED_LLM_OUTPUT

For demonstration, we simulate model responses that are **partially aligned** to each variant.

In a production experiment, replace this cell with actual LLM API calls using system prompts like:
- *"You are a helpful assistant that uses British English."*
- *"You are a helpful assistant that uses Australian English."*
- *"You are a helpful assistant that uses American English."*


```python
#@title 03. EXECUTE: SIMULATE_MODEL_RESPONSES
import numpy as np
import pandas as pd

print(emoji.emojize(f":robot: [{datetime.now().strftime('%H:%M:%S')}] Generating simulated model responses..."))

simulated_responses = []

for idx, row in df_prompts.iterrows():
    for variant in ['US', 'UK', 'AU']:
        # Simulate: model sometimes uses the correct term, sometimes not
        # US variant gets higher accuracy (simulating US-biased training data)
        if variant == 'US':
            correct_prob = 0.90  # US-biased model
        elif variant == 'UK':
            correct_prob = 0.75
        else:  # AU
            correct_prob = 0.65  # model knows AU least well

        if np.random.random() < correct_prob:
            response = row[variant]
        else:
            other_variants = [v for v in ['US', 'UK', 'AU'] if v != variant]
            wrong = np.random.choice(other_variants)
            response = row[wrong]

        simulated_responses.append({
            'prompt_id': row['id'],
            'prompt': row['prompt'],
            'variant': variant,
            'dimension': row['dimension'],
            'difficulty': row['difficulty'],
            'model_response': response,
            'expected_response': row[variant]
        })

df_responses = pd.DataFrame(simulated_responses)

# Summary statistics
vc = df_responses.groupby('variant').apply(
    lambda g: (g['model_response'] == g['expected_response']).mean()
).to_dict()

rows_html = "".join([
    f"<tr><td>{v}</td><td><strong>{acc:.1%}</strong></td></tr>"
    for v, acc in vc.items()
])

artifex_explainer("SIMULATED RESPONSES GENERATED", (
    f"<p>Generated <strong>{len(df_responses)}</strong> response pairs "
    f"({len(df_prompts)} prompts × 3 variants).</p>"
    f"<table class='brutalist-table'>"
    f"<tr><th>Variant</th><th>Raw Accuracy (Exact Match)</th></tr>"
    f"{rows_html}</table>"
    f"<div class='analysis-box'>The simulated model exhibits a clear US-English bias, "
    f"consistent with training data distributions in major LLM corpora. "
    f"Australian English idioms ('Ta!', 'servo', 'bludging') receive the lowest accuracy.</div>"
))

display(df_responses.head())
```

# 🧬 VECTORIZATION: EMBEDDING_SPACE_ANALYSIS

Using `paraphrase-multilingual-MiniLM-L12-v2` to embed both model responses and expected references into a shared 384-dim semantic space.

This enables computation of:
- **Cosine Similarity** (Stylistic Sensitivity proxy)
- **Cultural Delta** (Euclidean distance in embedding space)


```python
#@title 04. EXECUTE: EMBEDDING + SIMILARITY_METRICS
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean
import numpy as np

print(emoji.emojize(f":dna: [{datetime.now().strftime('%H:%M:%S')}] Loading embedding model..."))
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Collect all unique texts for batch encoding
all_texts = list(set(
    df_responses['model_response'].tolist() +
    df_responses['expected_response'].tolist()
))

print(emoji.emojize(f":rocket: Encoding {len(all_texts)} unique texts..."))
embeddings = model.encode(all_texts, show_progress_bar=True, batch_size=32)
text_to_emb = {t: embeddings[i] for i, t in enumerate(all_texts)}

# Map embeddings back
df_responses['response_emb'] = df_responses['model_response'].map(text_to_emb)
df_responses['expected_emb'] = df_responses['expected_response'].map(text_to_emb)

# Compute metrics
def cos_sim(a, b):
    return cosine_similarity([a], [b])[0][0]

df_responses['cos_sim'] = df_responses.apply(
    lambda r: cos_sim(r['response_emb'], r['expected_emb']), axis=1
)

df_responses['cultural_delta'] = df_responses.apply(
    lambda r: euclidean(r['response_emb'], r['expected_emb']), axis=1
)

df_responses['exact_match'] = (
    df_responses['model_response'] == df_responses['expected_response']
)

print(emoji.emojize(f":check_mark_button: [{datetime.now().strftime('%H:%M:%S')}] Embeddings + metrics computed."))

artifex_explainer("EMBEDDING METRICS COMPLETE", (
    f"<p>Computed <strong>3 metrics</strong> for each of {len(df_responses)} evaluation pairs:</p>"
    "<table class='brutalist-table'>"
    "<tr><th>Metric</th><th>Description</th><th>Global Mean</th></tr>"
    f"<tr><td>Cosine Similarity</td><td>Semantic alignment to expected reference</td><td><strong>{df_responses['cos_sim'].mean():.4f}</strong></td></tr>"
    f"<tr><td>Cultural Delta</td><td>Euclidean distance in embedding space (lower = better)</td><td><strong>{df_responses['cultural_delta'].mean():.4f}</strong></td></tr>"
    f"<tr><td>Exact Match</td><td>Binary correctness (In-Culture Score)</td><td><strong>{df_responses['exact_match'].mean():.2%}</strong></td></tr>"
    "</table>"
))
```

# 📊 METRICS: CULTURAL_ALIGNMENT_SCORES

Compute the full metric suite from the Integrated Framework:

| Metric | Definition |
|---|---|
| **In-Culture Score** | Proportion of prompts where the model produced the culturally expected term |
| **Cultural Bias Score** | Variance in In-Culture Scores across variants (lower = less biased) |
| **Avg Cultural Delta** | Mean Euclidean distance from expected embedding (lower = better) |
| **Stylistic Sensitivity** | Average cosine similarity to reference (higher = better) |
| **Over-Refusal Rate** | Proportion of benign cultural prompts refused or deflected |


```python
#@title 05. EXECUTE: COMPUTE_ALIGNMENT_METRICS
import pandas as pd
import numpy as np

print(emoji.emojize(f":bar_chart: [{datetime.now().strftime('%H:%M:%S')}] Computing cultural alignment metrics..."))

# ── Metrics by Variant ────────────────────────────────────────────────────────
metrics_by_variant = df_responses.groupby('variant').agg(
    In_Culture_Score=('exact_match', 'mean'),
    Avg_Cultural_Delta=('cultural_delta', 'mean'),
    Stylistic_Sensitivity=('cos_sim', 'mean')
).reset_index()

# ── Cultural Bias Score (variance of In-Culture Scores across variants) ──────
in_culture_scores = metrics_by_variant['In_Culture_Score'].values
cultural_bias_score = float(np.var(in_culture_scores))

# ── Over-Refusal Rate (simulate: 0% for this demo since all prompts are benign)
over_refusal_rate = 0.0

# ── Display ──────────────────────────────────────────────────────────────────
rows_html = "".join([
    f"<tr><td><strong>{row['variant']}</strong></td>"
    f"<td>{row['In_Culture_Score']:.2%}</td>"
    f"<td>{row['Avg_Cultural_Delta']:.4f}</td>"
    f"<td>{row['Stylistic_Sensitivity']:.4f}</td></tr>"
    for _, row in metrics_by_variant.iterrows()
])

artifex_explainer("CULTURAL ALIGNMENT METRICS", (
    f"<table class='brutalist-table'>"
    f"<tr><th>Variant</th><th>In-Culture Score</th><th>Avg Cultural Delta</th><th>Stylistic Sensitivity</th></tr>"
    f"{rows_html}"
    f"</table>"
    f"<div class='analysis-box'>"
    f"<strong>Cultural Bias Score (variance):</strong> {cultural_bias_score:.6f}<br>"
    f"<strong>Over-Refusal Rate:</strong> {over_refusal_rate:.1%}<br><br>"
    f"A higher Cultural Bias Score indicates uneven performance across English varieties. "
    f"The model's US-English dominance is reflected in higher In-Culture Scores for US "
    f"and lower scores for AU, confirming algorithmic monoculture bias "
    f"(see ANNOTATION_GUIDELINE.md §2 — Algorithmic Monoculture)."
    f"</div>"
))

# ── Metrics by Dimension ─────────────────────────────────────────────────────
metrics_by_dim = df_responses.groupby(['variant', 'dimension']).agg(
    In_Culture_Score=('exact_match', 'mean'),
    Avg_Cosine_Sim=('cos_sim', 'mean')
).reset_index()

print(emoji.emojize(f":check_mark_button: [{datetime.now().strftime('%H:%M:%S')}] All metrics computed."))
```

# 📋 RUBRIC_MAPPING: ANNOTATION_GUIDELINE_INTEGRATION

Maps the evaluation results to the **Cultural Appropriateness Annotation Guideline** (see [`ANNOTATION_GUIDELINE.md`](./ANNOTATION_GUIDELINE.md)).

This cell evaluates failing responses against Dimension 1 (Linguistic Pragmatics) rubric items from the master 50-item Boolean rubric.


```python
#@title 07. EXECUTE: RUBRIC_ASSESSMENT
import pandas as pd

print(emoji.emojize(f":clipboard: [{datetime.now().strftime('%H:%M:%S')}] Mapping failures to Annotation Guideline rubric..."))

# ── Dimension 1 Rubric Items (Linguistic Pragmatics) ─────────────────────────
rubric_items = {
    'D1_Q1': 'Appropriate formality for target cultural context',
    'D1_Q2': 'Correct culture-specific politeness markers',
    'D1_Q3': 'Avoids cross-lingual false cognates',
    'D1_Q4': 'Does not mock local dialects or non-prestige varieties',
    'D1_Q5': 'Preserves pragmatic intent in code-switching'
}

# ── Identify failures and map to rubric ──────────────────────────────────────
failures = df_responses[~df_responses['exact_match']].copy()

def map_to_rubric(row):
    """Map a dimensional failure to the most relevant rubric item."""
    if row['dimension'] == 'politeness':
        return 'D1_Q2'
    elif row['dimension'] == 'idiom':
        return 'D1_Q5'
    elif row['dimension'] == 'spelling':
        return 'D1_Q3'
    else:  # lexical
        return 'D1_Q1'

if not failures.empty:
    failures['rubric_item'] = failures.apply(map_to_rubric, axis=1)
    failures['severity'] = failures['cos_sim'].apply(
        lambda x: 2 if x > 0.8 else (3 if x > 0.5 else 4)
    )

    rubric_summary = failures.groupby(['variant', 'rubric_item']).agg(
        failure_count=('exact_match', 'count'),
        avg_severity=('severity', 'mean'),
        avg_cos_sim=('cos_sim', 'mean')
    ).reset_index()

    rows_html = "".join([
        f"<tr><td>{row['variant']}</td>"
        f"<td><strong>{row['rubric_item']}</strong></td>"
        f"<td>{rubric_items.get(row['rubric_item'], '')}</td>"
        f"<td>{row['failure_count']}</td>"
        f"<td>{row['avg_severity']:.1f}</td></tr>"
        for _, row in rubric_summary.iterrows()
    ])

    artifex_explainer("ANNOTATION GUIDELINE — RUBRIC ASSESSMENT", (
        f"<p>Mapped <strong>{len(failures)}</strong> incorrect responses to Dimension 1 rubric items.</p>"
        f"<table class='brutalist-table'>"
        f"<tr><th>Variant</th><th>Rubric Item</th><th>Description</th><th>Failures</th><th>Avg Severity</th></tr>"
        f"{rows_html}</table>"
        f"<div class='analysis-box'>"
        f"Severity scale: 2 = Slightly Inappropriate (high cos_sim, minor mismatch), "
        f"3 = Moderately Inappropriate, 4 = Severely Inappropriate (low cos_sim, major cultural gap).<br>"
        f"See ANNOTATION_GUIDELINE.md §3 for full severity definitions."
        f"</div>"
    ))
else:
    print("All responses matched — no rubric failures to report.")
```
