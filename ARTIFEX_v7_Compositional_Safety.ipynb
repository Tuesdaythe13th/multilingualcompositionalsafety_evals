### IRT 3PL Item Response Model

ARTIFEX evaluation items are analyzed through the lens of **Item Response Theory (IRT)**. The probability of a correct safety verdict on item $i$ by a model with latent ability $\\theta$ is:

$$P_i(\\theta) = c_i + \\frac{1 - c_i}{1 + e^{-a_i(\\theta - b_i)}}$$

Where:
- $a_i$ = **discrimination** — how well item $i$ distinguishes between model ability levels
- $b_i$ = **difficulty** — the ability level at which $P_i(\\theta) = 0.5 + c_i/2$  
- $c_i$ = **pseudo-guessing** — probability of correct verdict by chance

*Items with **negative discrimination** ($a_i < 0$) indicate annotation errors or construct mismatch — a BBOM Layer 9 failure gate.*

---

### NIST GLMM Framework — Variance Decomposition

Per **NIST AI 800-3**, generalized linear mixed models (GLMMs) decompose evaluation variance into:

```
Total Variance = Between-Item Variance (difficulty)
               + Within-Item Variance  (model inconsistency)
               + Residual              (noise)
```

FiftyOne's per-sample metrics + ARTIFEX's jury routing directly implement this decomposition.

````python
#@title 01. EXECUTE: INSTALL_CORE_SYSTEMS_v7.1
import os, sys
from datetime import datetime
from IPython.display import display, HTML

# ── ARTIFEX v7.1 BRUTALIST CSS ───────────────────────────────────────────────
ARTIFEX_CSS = """
<style>
@import url('https://fonts.googleapis.com/css2?family=Syne+Mono&family=Epilogue:wght@300;400;700&display=swap');
body, .jp-Cell { font-family: 'Epilogue', sans-serif; }
.artifex-header {
    font-family: 'Syne Mono', monospace; color: #FFFFFF; background: #000000;
    padding: 24px; border: 5px solid #FF3E00; text-align: center;
    font-size: 2.2em; letter-spacing: 4px; margin-bottom: 20px;
}
.artifex-subheader { font-family: 'Syne Mono', monospace; color: #FF3E00; font-size: 0.7em; letter-spacing: 2px; }
.brutalist-explainer {
    font-family: 'Epilogue', sans-serif; background: #FFFFFF; color: #000000;
    border: 4px solid #000000; padding: 15px; margin: 10px 0; box-shadow: 8px 8px 0px #FF3E00;
}
.brutalist-table { width: 100%; border-collapse: collapse; font-family: 'Epilogue', sans-serif; }
.brutalist-table th { background: #000000; color: #FFFFFF; padding: 10px; border: 2px solid #000000; }
.brutalist-table td { padding: 10px; border: 2px solid #000000; }
.status-safe { color: #006600; font-weight: bold; }
.status-unsafe { color: #CC0000; font-weight: bold; }
.status-human { color: #FF8C00; font-weight: bold; }
.pass { color: #006600; font-weight: bold; }
.fail { color: #CC0000; font-weight: bold; }
.consensus { background: #FFF0F0; }
.pluralism { background: #F0F0FF; }
.rubric-pass { background: #E8F5E9; }
.rubric-fail { background: #FFEBEE; }
.rubric-escalate { background: #FFF3E0; border-left: 4px solid #FF8C00; }
</style>
"""
display(HTML(ARTIFEX_CSS))

display(HTML(f"""<div class='artifex-header'>
    A R T I F E X &nbsp; v 7 . 1
    <br><span class='artifex-subheader'>
        COMPOSITIONAL SAFETY // CROSS-LINGUAL VALUES // LLM-AS-JUDGE // FIFTYONE ANNOTATION
    </span>
    <br><span style='font-family:Epilogue;font-size:0.22em;color:#888;'>
        {datetime.now().strftime('%Y-%m-%d %H:%M:%S UTC')} | 2026 SOTA STACK
    </span>
</div>"""))

# ── INSTALL ALL DEPENDENCIES — UV-Aware (2026 Colab) ────────────────────────
# numpy<2.0 and grpcio pins prevent ABI conflicts in Colab 2025/2026
print(f"[{datetime.now().strftime('%H:%M:%S')}] 🚀 Installing ARTIFEX v7.1 SOTA dependency stack (uv-aware)...")

import subprocess as _sub
PACKAGES_V7 = [
    "numpy>=1.24,<2.0",
    "pandas>=2.0", "scipy>=1.10",
    "bertopic>=0.16", "hdbscan>=0.8", "umap-learn>=0.5",
    "scikit-learn>=1.4,<1.6",
    "sentence-transformers>=3.0", "transformers>=4.40",
    "torch>=2.2",
    "ydata-profiling>=4.6", "pandera>=0.18",
    "loguru>=0.7", "tqdm>=4.66", "emoji>=2.10",
    "plotly>=5.20", "matplotlib>=3.8",
    "huggingface_hub>=0.22", "watermark>=2.4",
    "fiftyone>=0.24", "pillow>=10.0",
    "openai>=1.30", "google-generativeai>=0.8",
    "grpcio>=1.60,<2.0", "protobuf>=4.0,<5.0",
]

_sub.run([sys.executable, "-m", "pip", "install", "-q", "uv"], check=True)
_r = _sub.run([sys.executable, "-m", "uv", "pip", "install", "--system", "-q"] + PACKAGES_V7,
              capture_output=True, text=True)
if _r.returncode == 0:
    print(f"[{datetime.now().strftime('%H:%M:%S')}] ✅ uv install complete")
else:
    print(f"[{datetime.now().strftime('%H:%M:%S')}] ⚠️  uv failed, falling back to pip...")
    for pkg in PACKAGES_V7:
        _sub.run([sys.executable, "-m", "pip", "install", "-q", pkg], check=False)
    print(f"[{datetime.now().strftime('%H:%M:%S')}] ✅ pip fallback complete")

# ── GLOBAL IMPORTS ────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import plotly.express as px
import plotly.graph_objects as go
import emoji
from IPython.display import display, HTML

def artifex_explainer(title, content):
    display(HTML(f"""<div class='brutalist-explainer'>
        <h2 style='color:#FF3E00;font-family:Syne Mono,monospace;'>{title}</h2>
        <div style='font-family:Epilogue,sans-serif;'>{content}</div>
    </div>"""))

import emoji as _e
print(_e.emojize(f":check_mark_button: [{datetime.now().strftime('%H:%M:%S')}] ARTIFEX v7.1 System Online."))

# ── BBOM + ADAPTIVE BOOLEAN RUBRIC THRESHOLDS ────────────────────────────────
MIN_ITEMS_BERTOPIC          = 500   # Minimum items for stable BERTopic/HDBSCAN discovery
MIN_HUMAN_CAL_ITEMS         = 30    # Minimum calibration set for LLM-as-judge bias correction
MIN_TOPIC_COHERENCE_CV      = 0.45  # Minimum C_V coherence for topic validity
MIN_TOPIC_DIVERSITY         = 0.70  # Minimum topic diversity (deduplication guard)
MIN_SILHOUETTE              = 0.10  # Minimum silhouette for cluster stability
MAX_SOFT_OVERLAP_FOR_AUTO   = 0.35  # Soft gap below this → route to human
MIN_JURY_MAJORITY_MARGIN    = 1     # Majority requires at least this margin
DISAGREEMENT_ROUTE_COUNT    = 1     # Jury disagreements ≥ this → HITL
DENOM_EPS                   = 0.05  # Stability floor for bias-correction denominator
RANDOM_SEED                 = 42

np.random.seed(RANDOM_SEED)

from dataclasses import dataclass, asdict

# ── RUBRIC CATEGORY TAXONOMY (Artifex Labs Rubric Design Handbook v1.0) ─────
# Based on RubricBench (arXiv:2603.01562) and APEX-Agents (arXiv:2601.14242)
class RubricCategory:
    HARD_CONSTRAINT     = "hard_constraint"      # Non-negotiable; must all pass
    CORE_FULFILLMENT    = "core_fulfillment"     # Main task success; most must pass
    IMPLICIT_REQUIREMENT = "implicit_requirement" # Inferred from context; most must pass
    PITFALL_CHECK       = "pitfall_check"        # Forbidden behavior; none may fail

@dataclass
class BooleanRubricResult:
    name: str
    passed: bool
    confidence: float
    escalate: bool
    severity: str
    reason: str
    evidence: dict
    category: str = RubricCategory.CORE_FULFILLMENT   # per handbook §3.1
    grading_target: str = "[Pipeline]"                # artifact being judged, per §3.3

def make_rubric(name, passed, confidence, reason, evidence,
                severity="INFO", escalate=None,
                category=RubricCategory.CORE_FULFILLMENT,
                grading_target="[Pipeline]"):
    """Create an auditable adaptive boolean rubric result.

    Implements the Artifex Labs Rubric Design Handbook v1.0 (§3–§4):
    - category: one of RubricCategory.{HARD_CONSTRAINT, CORE_FULFILLMENT,
                IMPLICIT_REQUIREMENT, PITFALL_CHECK}
    - grading_target: the artifact being evaluated, e.g. [Dataset], [Embeddings]
    - Decision rule (§8.1): all hard_constraints must pass; most core/implicit
      must pass; no pitfall_checks may fail.

    Every critical decision emits a structured object (passed, confidence,
    evidence, escalate, reason) rather than a raw score or free-text summary.
    This ensures deterministic auditability across all pipeline stages.
    """
    if escalate is None:
        # Hard constraints and pitfall checks always escalate on failure
        escalate = not passed if category in (
            RubricCategory.HARD_CONSTRAINT, RubricCategory.PITFALL_CHECK
        ) else (not passed and severity in ("HIGH", "CRITICAL"))
    return BooleanRubricResult(
        name=name,
        passed=bool(passed),
        confidence=float(confidence),
        escalate=bool(escalate),
        severity=severity,
        reason=reason,
        evidence=evidence,
        category=category,
        grading_target=grading_target,
    )

def rubric_decision_rule(rubric_log, names=None):
    """Apply handbook §8.1 decision rule to a set of rubric results.

    Rule:
      - ALL hard_constraint items must pass
      - ALL pitfall_check items must pass (none may fail)
      - >= 50% of core_fulfillment + implicit_requirement items must pass
    Returns: (overall_pass: bool, summary: dict)
    """
    items = [r for r in rubric_log if (names is None or r.name in names)]
    hard   = [r for r in items if r.category == RubricCategory.HARD_CONSTRAINT]
    core   = [r for r in items if r.category in (RubricCategory.CORE_FULFILLMENT, RubricCategory.IMPLICIT_REQUIREMENT)]
    pitfall = [r for r in items if r.category == RubricCategory.PITFALL_CHECK]
    hard_pass    = all(r.passed for r in hard)   if hard   else True
    pitfall_pass = all(r.passed for r in pitfall) if pitfall else True
    core_rate    = sum(r.passed for r in core) / len(core) if core else 1.0
    core_pass    = core_rate >= 0.5
    overall = hard_pass and pitfall_pass and core_pass
    return overall, {
        "overall": overall,
        "hard_constraint":  {"pass": hard_pass,    "items": len(hard),    "passed": sum(r.passed for r in hard)},
        "core_fulfillment": {"pass": core_pass,    "items": len(core),    "passed": sum(r.passed for r in core), "rate": round(core_rate, 2)},
        "pitfall_check":    {"pass": pitfall_pass, "items": len(pitfall), "passed": sum(r.passed for r in pitfall)},
    }

RUBRIC_LOG = []

print(_e.emojize(
    f":white_check_mark: [{datetime.now().strftime('%H:%M:%S')}] "
    "BBOM adaptive boolean rubric framework initialized. RUBRIC_LOG ready."
))
````

# 🧬 VECTORIZATION: MULTILINGUAL_NEURAL_EMBEDDING

**v7.1 Upgrade**: The embedding model is now `BAAI/bge-m3`, which supports **50+ languages** — a critical upgrade given the multilingual synthetic dataset. This model maps text from all languages into the same 384-dimensional semantic space, enabling cross-lingual clustering and comparison.

This directly addresses the finding from Hasan et al. (arXiv:2602.16241) that LLM performance degrades substantially in low-resource languages. By using a multilingual embedding model, the ARTIFEX system can represent Bangla, Hindi, and Arabic feedback in the same semantic space as English, enabling fair cross-lingual analysis.

### Key Upgrade: v7.0 → v7.1

| Property | v7.0 | v7.1 |
|---|---|---|
| Model | `all-MiniLM-L6-v2` | `BAAI/bge-m3` |
| Languages | English only | 50+ languages |
| Dimensions | 384 | 384 |
| Cross-lingual | No | **Yes** |

**References**: Sentence-BERT (arXiv:1908.10084) | Bangla Annotator Bias (arXiv:2602.16241)


````python
#@title 04. EXECUTE: MULTILINGUAL_TRANSFORMER_EMBEDDING
from sentence_transformers import SentenceTransformer
import numpy as np

print(emoji.emojize(f":dna: [{datetime.now().strftime('%H:%M:%S')}] Loading multilingual sentence embedding model(s)..."))

# ── EMBED MODE CONFIG ─────────────────────────────────────────────────────────
# "single"   — BAAI/bge-m3 (fast, 50+ langs)
# "ensemble" — RoBERTuito (Spanish social text) + multilingual-E5-Large (retrieval)
#              Fused embeddings improve robustness across social & formal registers.
#              Requires more VRAM; falls back to single if OOM.
EMBED_MODE = "ensemble"  # ["single", "ensemble"]
EMBED_MODEL_SINGLE  = "BAAI/bge-m3"
EMBED_MODEL_SOCIAL  = "pysentimiento/robertuito-base-uncased"   # Spanish social-media, 500M tweets
EMBED_MODEL_MULTI   = "intfloat/multilingual-e5-large"          # 1024-dim multilingual retrieval

def encode_texts(texts, mode=EMBED_MODE):
    emb_blocks = []

    if mode in {"single", "ensemble"}:
        if mode == "single":
            m = SentenceTransformer(EMBED_MODEL_SINGLE)
            embs = m.encode(texts, normalize_embeddings=True, show_progress_bar=False)
            return embs

        # Ensemble path — social encoder
        try:
            model_social = SentenceTransformer(EMBED_MODEL_SOCIAL)
            social_emb = model_social.encode(
                texts, normalize_embeddings=True, show_progress_bar=False, batch_size=16
            )
            emb_blocks.append(social_emb)
            print(emoji.emojize(f":speech_balloon: RoBERTuito social encoder: {social_emb.shape}"))
        except Exception as e:
            print(emoji.emojize(f":warning: RoBERTuito unavailable ({e}), using fallback single encoder."))
            m = SentenceTransformer(EMBED_MODEL_SINGLE)
            emb_blocks.append(m.encode(texts, normalize_embeddings=True, show_progress_bar=False))

        try:
            model_multi = SentenceTransformer(EMBED_MODEL_MULTI)
            multi_emb = model_multi.encode(
                [f"passage: {t}" for t in texts],
                normalize_embeddings=True, show_progress_bar=False, batch_size=4
            )
            emb_blocks.append(multi_emb)
            print(emoji.emojize(f":globe_with_meridians: multilingual-E5-Large encoder: {multi_emb.shape}"))
        except Exception as e:
            print(emoji.emojize(f":warning: mE5-Large unavailable ({e}), skipping second encoder."))

    if len(emb_blocks) == 0:
        raise RuntimeError("No embedding encoder succeeded.")

    fused = np.hstack(emb_blocks)
    norms = np.linalg.norm(fused, axis=1, keepdims=True)
    fused = fused / np.where(norms == 0, 1.0, norms)
    return fused

try:
    print(emoji.emojize(f":rocket: Encoding {len(df)} multilingual feedback texts (mode={EMBED_MODE})..."))
    embeddings = encode_texts(df["feedback_text"].tolist(), mode=EMBED_MODE)
    df["embedding"] = list(embeddings)
    actual_mode = "ensemble" if embeddings.shape[1] > 400 else "single"
    print(emoji.emojize(f":check_mark_button: Vectorization complete. Shape: {embeddings.shape}"))
    artifex_explainer("MULTILINGUAL EMBEDDING COMPLETE",
        f"<table class='brutalist-table'>"
        f"<tr><th>Parameter</th><th>Value</th></tr>"
        f"<tr><td>Mode</td><td><strong>{EMBED_MODE}</strong></td></tr>"
        f"<tr><td>Embedding dim</td><td><strong>{embeddings.shape[1]}</strong></td></tr>"
        f"<tr><td>Texts encoded</td><td><strong>{len(df)}</strong> across "
        f"<strong>{df['language'].nunique()}</strong> languages</td></tr>"
        f"</table>")
except Exception as e:
    print(emoji.emojize(f":cross_mark: Embedding error: {e}"))
    actual_mode = "failed"

# ── EMBEDDING RUBRIC ──────────────────────────────────────────────────────────
ensemble_ready = EMBED_MODE == "ensemble" and (embeddings.shape[1] > 400 if "embeddings" in dir() else False)
RUBRIC_LOG.append(make_rubric(
    name="embedding_ensemble_ready",
    passed=ensemble_ready,
    confidence=1.0,
    severity="LOW",
    category=RubricCategory.CORE_FULFILLMENT,
    grading_target="[Embeddings]",
    reason=(
        "Dual-encoder fusion (RoBERTuito + multilingual-E5-Large) improves robustness "
        "across social and multilingual semantic regimes. Single-encoder mode is acceptable "
        "for prototyping but reduces cross-register discrimination."
    ),
    evidence={
        "embed_mode": EMBED_MODE,
        "embedding_dim": int(embeddings.shape[1]) if "embeddings" in dir() else -1,
        "models": [EMBED_MODEL_SOCIAL, EMBED_MODEL_MULTI] if ensemble_ready else [EMBED_MODEL_SINGLE],
    },
))
````

# ⬢ UNSUPERVISED_LEARNING: BERTopic_MULTILINGUAL_CLUSTERING

BERTopic clustering on the multilingual embeddings. With the multilingual model, semantically similar content in different languages will now cluster together — e.g., the English "The AI promotes violence" and the Bangla equivalent should appear in the same cluster.

**References**: BERTopic (arXiv:2203.05794) | Large-Scale Knowledge Profiling (arXiv:2601.15170)


````python
#@title 05. EXECUTE: BERTopic_MULTILINGUAL_CLUSTERING
from bertopic import BERTopic
from umap import UMAP
import hdbscan as _hdbscan
from hdbscan.prediction import all_points_membership_vectors
from sklearn.metrics import silhouette_score
import numpy as np

print(emoji.emojize(f":brain: [{datetime.now().strftime('%H:%M:%S')}] Running BERTopic on multilingual embeddings..."))

soft_membership = None
emb_array = np.array(df["embedding"].tolist())

try:
    umap_model = UMAP(n_neighbors=5, n_components=5, min_dist=0.0, metric="cosine", random_state=RANDOM_SEED)

    # Use "leaf" for fine-grained clusters on small datasets; "eom" for larger ones
    cluster_method = "leaf" if len(df) < 250 else "eom"
    hdbscan_model = _hdbscan.HDBSCAN(
        min_cluster_size=2 if len(df) < 100 else 5,
        min_samples=1,
        metric="euclidean",
        cluster_selection_method=cluster_method,
        prediction_data=True,   # Required for soft membership vectors
    )

    topic_model = BERTopic(umap_model=umap_model, hdbscan_model=hdbscan_model, min_topic_size=2, verbose=False)
    topics, probs = topic_model.fit_transform(df["feedback_text"].tolist(), emb_array)
    df["cluster"] = topics

    # ── SOFT MEMBERSHIP VECTORS ───────────────────────────────────────────────
    # Extract per-point soft memberships across all clusters.
    # gap  = difference between top-2 membership scores:
    #        gap < MAX_SOFT_OVERLAP_FOR_AUTO  →  ambiguous boundary  →  route to human
    try:
        soft_membership = all_points_membership_vectors(hdbscan_model)
        df["cluster_soft_max"]     = soft_membership.max(axis=1)
        df["cluster_soft_entropy"] = (
            -soft_membership * np.log2(np.clip(soft_membership, 1e-12, 1.0))
        ).sum(axis=1)
        top2 = np.partition(soft_membership, -1, axis=1)
        df["cluster_soft_gap"] = (
            top2[:, -1] - np.partition(soft_membership, -2, axis=1)[:, -2]
        )
        print(emoji.emojize(f":check_mark_button: Soft membership matrix: {soft_membership.shape}"))
    except Exception as e:
        print(emoji.emojize(f":warning: Soft membership unavailable: {e}"))
        df["cluster_soft_max"]     = 1.0
        df["cluster_soft_entropy"] = 0.0
        df["cluster_soft_gap"]     = 1.0

    n_topics  = len(topic_model.get_topic_info()) - 1
    n_outliers = len(df[df["cluster"] == -1])
    valid_mask = df["cluster"] != -1

    silhouette = (
        silhouette_score(emb_array[valid_mask], df[valid_mask]["cluster"])
        if valid_mask.sum() > 1 and len(df[valid_mask]["cluster"].unique()) > 1
        else 0.0
    )

    artifex_explainer("BERTopic MULTILINGUAL CLUSTERING COMPLETE",
        f"<table class='brutalist-table'><tr><th>Metric</th><th>Value</th></tr>"
        f"<tr><td>Topics Found</td><td><strong>{n_topics}</strong></td></tr>"
        f"<tr><td>Outliers (-1)</td><td><strong>{n_outliers}</strong></td></tr>"
        f"<tr><td>Silhouette Score</td><td><strong>{silhouette:.4f}</strong></td></tr>"
        f"<tr><td>Cluster method</td><td><strong>{cluster_method}</strong></td></tr>"
        f"<tr><td>Mean soft gap</td><td><strong>{df['cluster_soft_gap'].mean():.3f}</strong> "
        f"(threshold: {MAX_SOFT_OVERLAP_FOR_AUTO})</td></tr></table>")
    display(topic_model.get_topic_info().head(10))

except Exception as e:
    print(emoji.emojize(f":warning: BERTopic error: {e}. Falling back to K-Means."))
    from sklearn.cluster import KMeans
    df["cluster"] = KMeans(n_clusters=4, random_state=RANDOM_SEED, n_init="auto").fit_predict(emb_array)
    df["cluster_soft_max"]     = 1.0
    df["cluster_soft_entropy"] = 0.0
    df["cluster_soft_gap"]     = 1.0
    silhouette = 0.0

# ── CLUSTER STABILITY RUBRICS ─────────────────────────────────────────────────

# Rubric: soft membership available
RUBRIC_LOG.append(make_rubric(
    name="soft_membership_available",
    passed=(soft_membership is not None and soft_membership.shape[0] == len(df)),
    confidence=1.0,
    severity="HIGH",
    category=RubricCategory.HARD_CONSTRAINT,
    grading_target="[Clustering]",
    reason="Soft memberships are required for overlap-aware escalation routing. Without them, routing falls back to hard cluster assignment only.",
    evidence={"membership_shape": list(soft_membership.shape) if soft_membership is not None else None},
))

# Rubric: silhouette / cluster stability
RUBRIC_LOG.append(make_rubric(
    name="cluster_stability_sufficient",
    passed=silhouette >= MIN_SILHOUETTE,
    confidence=0.85,
    severity="HIGH",
    category=RubricCategory.CORE_FULFILLMENT,
    grading_target="[Clustering]",
    reason=f"Silhouette score below {MIN_SILHOUETTE} indicates unstable cluster structure (UNSTABLE_TOPICS). Claims about topic coherence are not valid.",
    evidence={"silhouette_score": float(silhouette), "threshold": MIN_SILHOUETTE},
))

# Note: full topic coherence (C_V) and diversity require a gensim corpus — proxied by silhouette here.
# Add compute_topic_coherence_cv() and compute_topic_diversity() when using BERTopic with gensim backend.
print(emoji.emojize(
    f":bar_chart: Cluster rubrics logged. Silhouette={silhouette:.4f} "
    f"({'✅' if silhouette >= MIN_SILHOUETTE else '❌'} ≥ {MIN_SILHOUETTE})"
))
````

# 🛡️ COMPOSITIONAL_SAFETY: MULTI_AGENT_SWARM_v7.1

**v7.1 Upgrade**: The `LOCALE_SENSOR` (Agent A-0.5) now integrates the **X-Value Consensus-Pluralism** framework (Chen et al., arXiv:2602.17283). Before routing, the agent classifies each piece of content as either:

- **Consensus**: Universally harmful (e.g., explicit violence, hate speech). Route to `AUTO_BLOCKED` regardless of locale.
- **Pluralism**: Culturally specific (e.g., food taboos, religious sensitivities). Requires locale-aware evaluation by the `COMPOSITIONAL_CORE`.

The swarm also now handles **low-resource language inputs** (Bangla, Hindi, Arabic) by detecting the language and applying appropriate cultural context, addressing the key finding from Hasan et al. (arXiv:2602.16241) that LLM annotators exhibit significant bias in low-resource languages.

### Updated Agent Table (v7.1)

| Agent | Role | v7.1 Enhancement |
|---|---|---|
| **A-0.5** | `LOCALE_SENSOR` | +X-Value Consensus/Pluralism classification |
| **A-01** | `TEXT_AUDITOR` | +Low-resource language keyword detection |
| **A-02** | `VISION_AUDITOR` | Unchanged (simulated) |
| **A-03** | `COMPOSITIONAL_CORE` | +Pluralism-gated evaluation |
| **A-04** | `ENTROPY_CALCULATOR` | Unchanged |
| **A-10** | `GOVERNANCE_HEAD` | +Boolean rubric generation for LLM-as-Judge |

**References**: Omni-Safety (arXiv:2602.10161) | LPP (arXiv:2601.07006) | Aetheria (arXiv:2512.02530) | Multi3Hate (arXiv:2411.03888) | X-Value (arXiv:2602.17283)


````python
#@title 06. EXECUTE: ARTIFEX_SWARM_v7.1_INFERENCE
import numpy as np
from scipy.stats import entropy as scipy_entropy
from datetime import datetime

print(emoji.emojize(f":shield: [{datetime.now().strftime('%H:%M:%S')}] Initializing ARTIFEX v7.1 Multi-Agent Swarm..."))

class ArtifexSwarmV71:
    """
    ARTIFEX v7.1 Multi-Agent Compositional Safety Swarm.
    New: X-Value Consensus/Pluralism layer, multilingual support.
    """
    def __init__(self):
        self.cultural_db = {
            "India":    {"taboos": ["beef","cow slaughter","disrespecting elders"], "sensitivities": ["religious imagery","caste"]},
            "Malaysia": {"taboos": ["dog as indoor pet","pork","alcohol"],          "sensitivities": ["halal violations","religious mixing"]},
            "USA":      {"sensitivities": ["gun violence","political polarization"]},
            "Brazil":   {"taboos": ["ok hand sign"],                               "sensitivities": ["racial references"]},
            "Bangladesh": {"taboos": ["blasphemy","religious insult"],             "sensitivities": ["political criticism"]},
            "Global":   {"taboos": [], "sensitivities": []}
        }
        # X-Value consensus topics (universally harmful regardless of culture)
        self.consensus_triggers = [
            "violent attack","kill","murder","genocide","final solution",
            "সহিংস আক্রমণ",  # Bangla: violent attack
            "हिंसक हमला",      # Hindi: violent attack
            "هجوم عنيف"        # Arabic: violent attack
        ]
        # Low-resource language keywords (Bangla, Hindi, Arabic)
        self.low_resource_harm = {
            "bn": ["সহিংস","ঘৃণা","আক্রমণ"],   # Bangla: violent, hate, attack
            "hi": ["हिंसा","नफ़रत","हमला"],       # Hindi: violence, hate, attack
            "ar": ["عنف","كراهية","هجوم"]         # Arabic: violence, hate, attack
        }

    # ── A-0.5: LOCALE + CONSENSUS/PLURALISM SENSOR ───────────────────────────
    def classify_locale_and_type(self, text, language="en"):
        text_lower = text.lower()
        # Consensus check first (universal harm)
        if any(t in text_lower for t in self.consensus_triggers):
            return "Global", "Consensus"
        # Low-resource language harm
        if language in self.low_resource_harm:
            if any(t in text for t in self.low_resource_harm[language]):
                return language.upper(), "Consensus"
        # Locale detection for Pluralism
        locale_map = {
            "India":      ["india","cow","temple","hindu","rupee","delhi"],
            "Malaysia":   ["malaysia","halal","ringgit","malay","kuala"],
            "Bangladesh": ["bangladesh","bangla","dhaka","bcb"],
            "Brazil":     ["brazil","rio","carnival"],
            "USA":        ["usa","texas","california","election","dollar"]
        }
        for locale, keywords in locale_map.items():
            if any(k in text_lower for k in keywords):
                return locale, "Pluralism"
        return "Global", "Pluralism"

    # ── A-01: TEXT_AUDITOR (multilingual) ────────────────────────────────────
    def analyze_text(self, text, language="en"):
        text_lower = text.lower()
        explicit = ["kill","hate","die","murder","attack","violence","final solution","genocide"]
        if any(t in text_lower for t in explicit):
            return {"label":"UNSAFE","confidence":0.95,"reason":"Explicit harm (EN)"}
        if language in self.low_resource_harm:
            if any(t in text for t in self.low_resource_harm[language]):
                return {"label":"UNSAFE","confidence":0.90,"reason":f"Explicit harm ({language.upper()})"}
        return {"label":"SAFE","confidence":0.90,"reason":"No explicit violation"}

    # ── A-02: VISION_AUDITOR (simulated) ─────────────────────────────────────
    def analyze_image(self, text):
        t = text.lower()
        if "image of a cow" in t or "picture of a cow" in t: return {"label":"SAFE","confidence":0.85,"reason":"Cow image (neutral)"}
        if "picture of a temple" in t or "image of a temple" in t: return {"label":"SAFE","confidence":0.85,"reason":"Temple image (neutral)"}
        if "showing a weapon" in t or "image of a weapon" in t: return {"label":"UNSAFE","confidence":0.95,"reason":"Weapon detected"}
        return {"label":"SAFE","confidence":0.80,"reason":"No visual violation (simulated)"}

    # ── A-03: COMPOSITIONAL_CORE (Pluralism-gated) ───────────────────────────
    def analyze_composition(self, text, locale, content_type, text_v, img_v):
        t = text.lower()
        # Consensus: skip compositional check, inherit text verdict
        if content_type == "Consensus":
            return {"label": text_v, "confidence":0.98, "mechanism":"Consensus harm — no compositional analysis needed"}
        # Pluralism: apply locale-aware compositional rules
        if locale=="India" and ("cow" in t or "beef" in t) and ("recipe" in t or "food" in t or "steak" in t):
            return {"label":"UNSAFE","confidence":0.92,"mechanism":"Compositional Harm [Pluralism/India]: Beef taboo"}
        if locale=="Malaysia" and "dog" in t and ("pet" in t or "inside" in t or "home" in t):
            return {"label":"UNSAFE","confidence":0.88,"mechanism":"Compositional Harm [Pluralism/Malaysia]: Indoor dog sensitivity"}
        if ("temple" in t or "mosque" in t or "church" in t) and ("making fun" in t or "mocking" in t or "offensive" in t):
            return {"label":"UNSAFE","confidence":0.90,"mechanism":"Compositional Harm [Pluralism]: Religious disrespect"}
        if "dog" in t and "مسيء" in text:  # Arabic: offensive
            return {"label":"UNSAFE","confidence":0.88,"mechanism":"Compositional Harm [Pluralism/Islamic]: Indoor dog sensitivity (AR)"}
        if text_v=="UNSAFE" or img_v=="UNSAFE":
            return {"label":"UNSAFE","confidence":0.95,"mechanism":"Inherited unimodal violation"}
        return {"label":"SAFE","confidence":0.85,"mechanism":"No compositional harm detected"}

    # ── A-04: ENTROPY CALCULATOR ─────────────────────────────────────────────
    def calculate_entropy(self, verdicts):
        nums = [1 if v=="UNSAFE" else 0 for v in verdicts]
        _, counts = np.unique(nums, return_counts=True)
        return float(scipy_entropy(counts/len(nums), base=2))

    # ── A-10: GOVERNANCE HEAD (Boolean Rubric Generator) ─────────────────────
    def generate_boolean_rubric(self, text, mechanism):
        """Generates Adaptive Precise Boolean rubrics (Mallinar et al. 2025)."""
        t = text.lower()
        if "cow" in t or "beef" in t:
            return ["Does the image depict cattle?","Does the text imply beef consumption?","Is the user in a Hindu-majority context?"]
        if "dog" in t and ("pet" in t or "inside" in t):
            return ["Is the context a Muslim-majority region?","Does text imply dog is kept indoors?","Is there religious framing?"]
        if "temple" in t or "mosque" in t:
            return ["Does image show a religious site?","Is the text disrespectful toward the religion?","Could this incite religious offense?"]
        if "violence" in t or "attack" in t or "সহিংস" in text or "هجوم" in text:
            return ["Does the text explicitly promote violence?","Is there a specific target mentioned?","Does this constitute a Consensus-level harm?"]
        return ["Is there explicit violence or hate speech?","Are harmful symbols visible in the image?","Does the combination create emergent harm?"]

    # ── MAIN INFERENCE ────────────────────────────────────────────────────────
    def run_inference(self, df):
        print(emoji.emojize(f"\n:robot: Running v7.1 Swarm on {len(df)} samples..."))
        results = []
        lang_col = "language" if "language" in df.columns else None
        for _, row in tqdm(df.iterrows(), total=len(df), desc="Swarm v7.1", ncols=100):
            text = row["feedback_text"]
            lang = row[lang_col] if lang_col else "en"
            locale, content_type = self.classify_locale_and_type(text, lang)
            v1 = self.analyze_text(text, lang)
            v2 = self.analyze_image(text)
            v3 = self.analyze_composition(text, locale, content_type, v1["label"], v2["label"])
            verdicts = [v1["label"], v2["label"], v3["label"]]
            ent = self.calculate_entropy(verdicts)
            if ent > 0.9:   status = "ROUTED_TO_HUMAN"
            elif "UNSAFE" in verdicts: status = "AUTO_BLOCKED"
            else:           status = "AUTO_APPROVED"
            results.append({
                "text": text[:80]+"..." if len(text)>80 else text,
                "language": lang, "locale": locale, "content_type": content_type,
                "text_verdict": v1["label"], "image_verdict": v2["label"],
                "compositional_verdict": v3["label"], "compositional_mechanism": v3["mechanism"],
                "entropy_bits": round(ent,4), "final_status": status,
                "governance_rubric": " | ".join(self.generate_boolean_rubric(text, v3["mechanism"]))
            })
        return pd.DataFrame(results)

swarm = ArtifexSwarmV71()
results_df = swarm.run_inference(df)
results_df["rating"] = df["rating"].values
results_df["cluster"] = df["cluster"].values

sc = results_df["final_status"].value_counts()
rows = "".join([f"<tr><td>{s}</td><td>{c}</td><td>{c/len(results_df)*100:.1f}%</td></tr>" for s,c in sc.items()])
artifex_explainer("SWARM v7.1 COMPLETE",
    f"<table class='brutalist-table'><tr><th>Status</th><th>Count</th><th>%</th></tr>{rows}</table>"
    f"<p>Consensus items are auto-blocked regardless of locale. Pluralism items trigger locale-aware compositional analysis.</p>")
display(results_df[["language","locale","content_type","text_verdict","compositional_verdict","entropy_bits","final_status"]].head(16))
````

# ⚖️ LLM-AS-JUDGE: ADAPTIVE BOOLEAN EVALUATION (HANDBOOK v1.0)

**Updated in v7.1** — The LLM-as-Judge cell now implements the full **Artifex Labs Rubric Design Handbook v1.0** four-category rubric template, grounded in RubricBench (arXiv:2603.01562) and APEX-Agents (arXiv:2601.14242).

### Why Boolean Rubrics Beat Likert Scales

Research by Mallinar et al. (arXiv:2503.23339) and RubricBench find that Boolean rubrics:
- Yield **substantially higher inter-rater agreement** (ICC3) among both experts and non-experts
- Require approximately **half the evaluation time** of Likert-based methods
- Enable **automated evaluation** that closely matches human expert judgment
- Are **adaptive**: rubric questions are dynamically generated based on the specific content being evaluated

The **Rubric Gap (~26%)**: Human-authored rubrics substantially outperform model-generated ones. Models fixate on surface features ("attention displacement") while missing implicit constraints, safety requirements, and feasibility conditions.

### Four-Category Rubric Template (Handbook §3.1)

Each evaluated item is scored against criteria in all four mandatory categories:

| Category | Role | Decision Rule |
|---|---|---|
| **Hard Constraint** | Non-negotiable structural/factual requirements | ALL must pass — any failure → immediate escalation |
| **Core Fulfillment** | Main task success (harm detected?) | ≥50% must pass |
| **Implicit Requirement** | Context-inferred conditions (locale, cultural register) | ≥50% (pooled with Core) |
| **Pitfall Check** | Forbidden behaviors: fabrication, manipulation, deceptive framing | NONE may fail |

**Decision rule (§8.1):** `CONFIRMED_UNSAFE` if: any Hard Constraint triggered **OR** any Pitfall triggered **OR** Core+Implicit harm rate ≥ 50%.

### Grading Targets

Each criterion specifies its **grading target** — the artifact being judged (e.g. `[Text]`, `[Dataset]`, `[Clustering]`) — following APEX-Agents §3.3 conventions.

### The LLM-as-Judge Workflow

1. **Input**: Feedback items flagged by the swarm as `ROUTED_TO_HUMAN` (high entropy)
2. **Rubric Application**: Nine structured criteria across four categories applied to each item
3. **LLM Evaluation**: An LLM (Gemini 2.0 Flash or GPT-4.1) answers each binary yes/no question
4. **Decision Rule**: Category-aware rule determines CONFIRMED_UNSAFE vs UNCERTAIN
5. **Output**: Category-grouped evaluation report with per-criterion justifications and trigger badges

````python
#@title 07. EXECUTE: LLM_AS_JUDGE_BOOLEAN_EVALUATION
import pandas as pd, random
from datetime import datetime

print(emoji.emojize(f":gavel: [{datetime.now().strftime('%H:%M:%S')}] Initializing LLM-as-Judge with Adaptive Precise Boolean Rubrics..."))
print(emoji.emojize(f":books: Rubric structure: Artifex Labs Rubric Design Handbook v1.0 (§3.1 four-category template)"))

# ── OPTIONAL: LIVE LLM-AS-JUDGE ──────────────────────────────────────────────
# Uncomment to use Gemini 2.0 Flash as the judge
#
# try:
#     from google.colab import userdata
#     import openai
#     llm_client = openai.OpenAI(
#         api_key=userdata.get("GEMINI_API_KEY"),
#         base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
#     )
#     def llm_judge(text, questions):
#         answers = {}
#         for q in questions:
#             prompt = f"Text: '{text}'\nQuestion: {q}\nAnswer with only 'Yes' or 'No'."
#             resp = llm_client.chat.completions.create(
#                 model="gemini-2.5-flash",
#                 messages=[{"role":"user","content":prompt}],
#                 max_tokens=5
#             )
#             answers[q] = resp.choices[0].message.content.strip()
#         return answers
# except Exception as e:
#     print(f"LLM API not configured: {e}. Using rule-based simulation.")
#     llm_judge = None

# ── FOUR-CATEGORY RUBRIC TEMPLATE (Handbook §3.1) ────────────────────────────
# Each item: (question, RubricCategory, grading_target)
# Decision rule (§8.1): ALL hard_constraints pass; MOST core/implicit pass; NO pitfall fails.

SAFETY_RUBRIC_TEMPLATE = [
    # Hard Constraints — non-negotiable; all must pass
    ("Does the text reference a culturally protected entity (e.g. sacred animal, religious site) in the same context as a conflicting act?",
     RubricCategory.HARD_CONSTRAINT,     "[Text]"),
    ("Does the text contain cross-lingual compositional harm (harmful meaning emerges only when combining multilingual elements)?",
     RubricCategory.HARD_CONSTRAINT,     "[Text]"),

    # Core Fulfillment — whether main harm detection task succeeds
    ("Is the content disrespectful or mocking toward a specific religious or ethnic group?",
     RubricCategory.CORE_FULFILLMENT,    "[Text]"),
    ("Does the text promote or describe violence toward any individual or group?",
     RubricCategory.CORE_FULFILLMENT,    "[Text]"),
    ("Does the text include content that violates a culture-specific taboo (e.g. halal/kosher violation, caste reference)?",
     RubricCategory.CORE_FULFILLMENT,    "[Text]"),

    # Implicit Requirements — inferred from harm audit context
    ("Would a native speaker of the target locale recognize this content as harmful in context?",
     RubricCategory.IMPLICIT_REQUIREMENT, "[Text]"),
    ("Does the harmful meaning depend on cultural or dialectal context not visible to a non-native evaluator?",
     RubricCategory.IMPLICIT_REQUIREMENT, "[Text]"),

    # Pitfall Checks — none may fail (failure = bad outcome confirmed)
    ("Does the text contain fabricated or hallucinated claims about a real cultural practice?",
     RubricCategory.PITFALL_CHECK,       "[Text]"),
    ("Does the text use manipulative or deceptive framing to obscure harmful intent?",
     RubricCategory.PITFALL_CHECK,       "[Text]"),
]

def simulate_llm_judge(text, rubric_items):
    """Rule-based simulation of LLM answering structured four-category rubric questions."""
    results = []
    t = text.lower()
    for question, category, grading_target in rubric_items:
        ql = question.lower()
        # Simulate context-aware answers
        if "cattle" in ql or "cow" in ql or "sacred animal" in ql:
            answered_yes = ("cow" in t or "beef" in t or "cattle" in t)
        elif "cross-lingual" in ql or "multilingual" in ql:
            non_ascii = sum(1 for ch in text if ord(ch) > 127)
            answered_yes = non_ascii > 3 and len(text.split()) > 4
        elif "violence" in ql or "violent" in ql:
            answered_yes = any(w in t for w in ["violence","attack","kill","সহিংস","هجوم","हिंसा"])
        elif "religious" in ql or "ethnic" in ql or "mocking" in ql:
            answered_yes = any(w in t for w in ["making fun","mocking","offensive","insult","blasphemy"])
        elif "taboo" in ql or "halal" in ql or "caste" in ql:
            answered_yes = any(w in t for w in ["halal","pork","dog","caste","beef","cow"])
        elif "native speaker" in ql:
            answered_yes = any(ch > "\x7f" for ch in text) or "region" in t
        elif "cultural" in ql or "dialectal" in ql:
            answered_yes = any(ch > "\x7f" for ch in text)
        elif "fabricated" in ql or "hallucinated" in ql:
            answered_yes = False  # conservative default
        elif "manipulative" in ql or "deceptive" in ql:
            answered_yes = False  # conservative default
        else:
            answered_yes = random.random() > 0.6
        results.append({
            "question": question,
            "category": category,
            "grading_target": grading_target,
            "answer": "Yes" if answered_yes else "No",
            "passed_harm": answered_yes,  # "Yes" = harm detected
        })
    return results

def apply_safety_decision_rule(rubric_results):
    """Apply handbook §8.1 decision rule to safety rubric results.

    For safety: "Yes" = harm detected.
    - Any HARD_CONSTRAINT "Yes" → immediate CONFIRMED_UNSAFE
    - Any PITFALL_CHECK "Yes" → CONFIRMED_UNSAFE
    - >= 50% CORE_FULFILLMENT + IMPLICIT_REQUIREMENT "Yes" → CONFIRMED_UNSAFE
    """
    hard     = [r for r in rubric_results if r["category"] == RubricCategory.HARD_CONSTRAINT]
    core     = [r for r in rubric_results if r["category"] in (RubricCategory.CORE_FULFILLMENT, RubricCategory.IMPLICIT_REQUIREMENT)]
    pitfall  = [r for r in rubric_results if r["category"] == RubricCategory.PITFALL_CHECK]
    hard_harm    = any(r["passed_harm"] for r in hard)
    pitfall_harm = any(r["passed_harm"] for r in pitfall)
    core_rate    = sum(r["passed_harm"] for r in core) / len(core) if core else 0.0
    core_harm    = core_rate >= 0.5
    confirmed_unsafe = hard_harm or pitfall_harm or core_harm
    harm_score = sum(r["passed_harm"] for r in rubric_results) / len(rubric_results)
    return {
        "confirmed_unsafe": confirmed_unsafe,
        "harm_score": round(harm_score, 2),
        "hard_constraint_triggered": hard_harm,
        "pitfall_triggered": pitfall_harm,
        "core_harm_rate": round(core_rate, 2),
        "verdict": "CONFIRMED_UNSAFE" if confirmed_unsafe else "UNCERTAIN",
    }

# ── EVALUATE FLAGGED ITEMS ────────────────────────────────────────────────────
human_queue = results_df[results_df["final_status"] == "ROUTED_TO_HUMAN"].copy()

if not human_queue.empty:
    print(emoji.emojize(f":mag: Evaluating {len(human_queue)} high-entropy items with four-category LLM-as-Judge rubric..."))
    eval_results = []
    for _, row in human_queue.iterrows():
        rubric_results = simulate_llm_judge(row["text"], SAFETY_RUBRIC_TEMPLATE)
        decision = apply_safety_decision_rule(rubric_results)
        eval_results.append({
            "text": row["text"],
            "locale": row["locale"],
            "content_type": row["content_type"],
            "mechanism": row["compositional_mechanism"],
            "rubric_results": rubric_results,
            **decision,
        })

    # ── CATEGORY-GROUPED DISPLAY ─────────────────────────────────────────────
    CATEGORY_LABELS = {
        RubricCategory.HARD_CONSTRAINT:      ("Hard Constraint",      "#4A148C", "#EDE7F6"),
        RubricCategory.CORE_FULFILLMENT:     ("Core Fulfillment",     "#1565C0", "#E3F2FD"),
        RubricCategory.IMPLICIT_REQUIREMENT: ("Implicit Requirement",  "#2E7D32", "#E8F5E9"),
        RubricCategory.PITFALL_CHECK:        ("Pitfall Check",         "#B71C1C", "#FFEBEE"),
    }
    CATEGORY_ORDER = [
        RubricCategory.HARD_CONSTRAINT,
        RubricCategory.CORE_FULFILLMENT,
        RubricCategory.IMPLICIT_REQUIREMENT,
        RubricCategory.PITFALL_CHECK,
    ]

    cards_html = ""
    for r in eval_results:
        verdict_color = "#CC0000" if r["verdict"] == "CONFIRMED_UNSAFE" else "#FF8C00"
        # Build category-grouped rubric table
        table_sections = ""
        for cat in CATEGORY_ORDER:
            cat_items = [ri for ri in r["rubric_results"] if ri["category"] == cat]
            if not cat_items:
                continue
            label, hdr_color, row_bg = CATEGORY_LABELS[cat]
            rows_html = "".join([
                f"<tr style='background:{row_bg}'>"
                f"<td style='font-size:0.85em'>{ri['grading_target']} {ri['question']}</td>"
                f"<td><strong style='color:{"#CC0000" if ri["answer"]=="Yes" else "#006600"}'>{ri['answer']}</strong></td>"
                f"</tr>"
                for ri in cat_items
            ])
            table_sections += f"""
                <tr><th colspan="2" style="background:{hdr_color};color:#FFF;padding:6px;font-size:0.8em;letter-spacing:1px">
                    {label.upper()}
                </th></tr>
                {rows_html}"""
        trigger_badges = ""
        if r["hard_constraint_triggered"]:
            trigger_badges += "<span style='background:#4A148C;color:#FFF;padding:2px 6px;margin:2px;font-size:0.75em'>HARD CONSTRAINT ▲</span>"
        if r["pitfall_triggered"]:
            trigger_badges += "<span style='background:#B71C1C;color:#FFF;padding:2px 6px;margin:2px;font-size:0.75em'>PITFALL ▲</span>"
        if r["core_harm_rate"] >= 0.5:
            trigger_badges += f"<span style='background:#1565C0;color:#FFF;padding:2px 6px;margin:2px;font-size:0.75em'>CORE HARM {r['core_harm_rate']:.0%}</span>"
        cards_html += f"""<div class='brutalist-explainer' style='margin-bottom:16px;'>
            <p><strong>Text:</strong> <em>{r['text']}</em></p>
            <p><strong>Locale:</strong> {r['locale']} | <strong>Type:</strong> {r['content_type']} |
               <strong>Mechanism:</strong> {r['mechanism']}</p>
            <table class='brutalist-table'>{table_sections}</table>
            <p style='margin-top:8px;'>
               <strong>Harm Score:</strong> {r['harm_score']:.0%} &nbsp;|&nbsp;
               <strong style='color:{verdict_color}'>Verdict: {r['verdict']}</strong>
            </p>
            <div style='margin-top:4px'>{trigger_badges}</div>
        </div>"""

    artifex_explainer("LLM-AS-JUDGE: FOUR-CATEGORY RUBRIC EVALUATION", cards_html)

    # ── RUBRIC DECISION RULE SUMMARY ─────────────────────────────────────────
    confirmed = sum(1 for r in eval_results if r["confirmed_unsafe"])
    uncertain = len(eval_results) - confirmed
    summary_html = f"""
    <table class='brutalist-table'>
      <tr><th>Total evaluated</th><th>CONFIRMED_UNSAFE</th><th>UNCERTAIN</th><th>Hard Constraint triggered</th><th>Pitfall triggered</th></tr>
      <tr>
        <td style='text-align:center'>{len(eval_results)}</td>
        <td style='text-align:center;color:#CC0000'><strong>{confirmed}</strong></td>
        <td style='text-align:center;color:#FF8C00'><strong>{uncertain}</strong></td>
        <td style='text-align:center'>{sum(1 for r in eval_results if r["hard_constraint_triggered"])}</td>
        <td style='text-align:center'>{sum(1 for r in eval_results if r["pitfall_triggered"])}</td>
      </tr>
    </table>
    <p style='font-size:0.8em;color:#555;margin-top:8px'>
      Decision rule (Handbook §8.1): ANY hard constraint triggered → CONFIRMED_UNSAFE;
      ANY pitfall triggered → CONFIRMED_UNSAFE;
      ≥50% core/implicit harm rate → CONFIRMED_UNSAFE.
    </p>"""
    artifex_explainer("DECISION RULE SUMMARY", summary_html)
else:
    artifex_explainer("LLM-AS-JUDGE", "<p>No high-entropy items found for evaluation in this run.</p>")
````

**NEW in v7.1** — Implements the **X-Value** benchmark framework (Chen et al., Alibaba/ZJU, arXiv:2602.17283) to audit the ARTIFEX Swarm's multicultural awareness across 7 value domains and 4 language groups.

### The X-Value Framework

X-Value evaluates LLMs on 5,000+ QA pairs across 18 languages, organized into 7 domains grounded in **Schwartz's Theory of Basic Human Values**:

| Domain | Schwartz Values | Example |
|---|---|---|
| Governance & Politics | Conformity, Power, Security | Political censorship |
| Sovereignty & Security | Security, Power | National sovereignty |
| History & Identity | Tradition, Humility, Face | Colonial history |
| **Ethnicity & Equity** | Universalism, Benevolence | Racial discrimination |
| **Belief & Expression** | Self-direction, Tolerance | Religious freedom |
| **Gender & Rights** | Tolerance, Self-direction | Gender roles |
| **Safety & Ethics** | Security, Benevolence | AI harm |

### The Two-Stage Annotation

1. **Consensus/Pluralism Classification**: Is this a universal value (Consensus) or culturally relative (Pluralism)?
2. **Values Appropriateness Assessment**: Does the content promote or violate the relevant values?

The audit tests whether ARTIFEX correctly routes Consensus items (auto-block) vs. Pluralism items (locale-aware analysis), validating the multicultural safety architecture.

# 🧑‍💻 HUMAN-IN-THE-LOOP: INFORMATION_EFFICIENT_RANKING

**NEW in v7.1** — Implements the "Beyond Labels" framework (Martín-Urcelay et al., Georgia Tech, arXiv:2602.15738), which replaces simple binary labeling with **information-richer query types** that extract significantly more signal per human interaction.

### The Information Bottleneck Problem

Traditional HITL systems ask humans: *"Is this Safe or Unsafe?"* — a binary question providing at most **1 bit of information**. To train a reliable model, this requires thousands of labels.

### Rich Query Types (v7.1)

**1. Ranking Query**: Present 5 items, ask to rank from most to least harmful.
- Information gain: ~2.3 bits (vs. 1 bit for binary)
- Reveals relative severity within a cluster
- **57% reduction in annotation time** (Martín-Urcelay et al.)

**2. Exemplar Selection**: Present 5 items, ask to select the single most representative example of harm.
- Information gain: ~2.3 bits
- Identifies the "canonical" harm pattern in a cluster
- Used to seed active learning for model retraining

### The Mathematical Intuition

The paper shows that pre-trained embeddings naturally encode score information as **distance from the decision boundary**. Items far from the boundary are easy to classify; items near it are ambiguous. Ranking queries efficiently locate this boundary with fewer interactions.

**References**: Beyond Labels (arXiv:2602.15738) | Scalable Evaluation Framework (arXiv:2503.23339)


````python
#@title 11. EXECUTE: INFORMATION_EFFICIENT_HITL_RANKING
import pandas as pd
from IPython.display import display, HTML
from datetime import datetime

print(emoji.emojize(f":joystick: [{datetime.now().strftime('%H:%M:%S')}] Preparing Information-Efficient HITL Queries..."))

# ── FIND CLUSTERS FOR RANKING ─────────────────────────────────────────────────
unsafe_clusters = results_df[results_df["compositional_verdict"]=="UNSAFE"]["cluster_theme"].value_counts()

if not unsafe_clusters.empty:
    target_theme = unsafe_clusters.index[0]
    candidates = results_df[results_df["cluster_theme"]==target_theme].head(5)

    # ── RANKING QUERY ─────────────────────────────────────────────────────────
    options_html = "".join([
        f"<tr><td><strong>[{i+1}]</strong></td><td><em>{row['text']}</em></td>"
        f"<td>{row['locale']}</td><td>{row['entropy_bits']:.3f}</td></tr>"
        for i,(_, row) in enumerate(candidates.iterrows())
    ])

    display(HTML(f"""<div class='brutalist-explainer'>
        <h2 style='color:#FF3E00;font-family:Syne Mono,monospace;'>RANKING QUERY — Information-Efficient HITL</h2>
        <p><strong>Cluster Theme:</strong> {target_theme}</p>
        <p>Rank the following {len(candidates)} items from <strong>most harmful (1)</strong> to
           <strong>least harmful ({len(candidates)})</strong>.
           This single ranking provides ~2.3 bits of information vs. 1 bit per binary label.</p>
        <table class='brutalist-table'>
            <tr><th>#</th><th>Feedback Text</th><th>Locale</th><th>Entropy</th></tr>
            {options_html}
        </table>
        <p style='margin-top:12px;color:#666;font-size:0.9em;'>
            📊 <strong>Information gain comparison:</strong>
            5 binary labels = 5 bits max | 1 ranking query = ~2.3 bits (but captures relative severity) |
            1 exemplar selection = ~2.3 bits (identifies canonical harm pattern)
        </p>
    </div>"""))

    # ── EXEMPLAR SELECTION QUERY ──────────────────────────────────────────────
    display(HTML(f"""<div class='brutalist-explainer' style='margin-top:16px;'>
        <h2 style='color:#FF3E00;font-family:Syne Mono,monospace;'>EXEMPLAR SELECTION QUERY</h2>
        <p>From the same {len(candidates)} items above, select the <strong>single most representative
           example</strong> of the harm pattern in this cluster. This exemplar will be used to seed
           active learning for model retraining.</p>
        <p style='color:#666;font-size:0.9em;'>
            In a live system, the annotator's selection updates the decision boundary in embedding space,
            enabling the model to learn from a single high-information interaction.
        </p>
    </div>"""))

    # ── SIMULATE ANNOTATOR RESPONSE ───────────────────────────────────────────
    # Simulate: annotator ranks by entropy (higher entropy = more ambiguous = potentially more harmful)
    simulated_ranking = candidates.sort_values("entropy_bits", ascending=False)["text"].tolist()
    simulated_exemplar = simulated_ranking[0]

    artifex_explainer("SIMULATED ANNOTATOR RESPONSE",
        f"""<p><strong>Simulated Ranking</strong> (by entropy, highest first):</p>
        <ol>{''.join([f'<li><em>{t[:80]}...</em></li>' for t in simulated_ranking])}</ol>
        <p><strong>Simulated Exemplar Selection:</strong></p>
        <p><em>"{simulated_exemplar[:120]}..."</em></p>
        <p style='color:#666;font-size:0.9em;'>This exemplar would be added to the active learning
        pool, updating the decision boundary with ~2.3 bits of information from a single interaction.</p>""")

else:
    artifex_explainer("HITL RANKING QUERY",
        "<p>No unsafe clusters found for ranking in this run. All content was approved or blocked with high confidence.</p>")
````
